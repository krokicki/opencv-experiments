```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def show(cv_image):
    img = cv2.cvtColor(cv_image, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
# Generate a dictionary of 4x4 markers for printing 
dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_50)
for i in range(0,50):
    img = cv2.aruco.drawMarker(dictionary, i, 1024)
    cv2.imwrite('dict4x4/aruco_4x4_50_%04d.png'%i, img)
```

```python
image_filenames = [
    'aruco_test_21.jpg',
    'aruco_test_22.jpg',
    'aruco_test_23.jpg',
    'aruco_test_24.jpg'
]

def load_images():
    # Read some images
    images = []
    for filename in image_filenames:
        img = cv2.imread(filename)
        images.append(img)
    return images

# plot the raw images
images = load_images()
cols = 2
rows = (len(images) / cols) + 1
for i in range(len(images)):
    plt.title(image_filenames[i])
    show(images[i])
```

```python
dictionary = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_50)

for img in images:
    img2 = img.copy()
    res = cv2.aruco.detectMarkers(img2, dictionary)
    cv2.aruco.drawDetectedMarkers(img2, res[0], res[1])
    show(img2)
    
```
